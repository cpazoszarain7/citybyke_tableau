# Section 1: Merge Dataframes

```python
#Create List of Months and Years to Cover
Months = ['01','02','03','04','05','06','07','08','09','10','11','12']
Years = ['2020','2021']
Column_Names = ['tripduration', 'starttime', 'stoptime', 'start station id',
       'start station name', 'start station latitude',
       'start station longitude', 'end station id', 'end station name',
       'end station latitude', 'end station longitude', 'bikeid', 'usertype',
       'birth year', 'gender']
main_df = pd.DataFrame(columns=Column_Names)
main_df.head()
```

```python
for x in Years:
    for y in Months:
        if ((x == '2021') and (int(y) >=5)):
            break
        else:
            #Build name of csv file
            file = 'JC-'+x+y+'-citibike-tripdata.csv'
            url = os.path.join('..','Raw_Data',file)            
            current_df = pd.read_csv(url)
            main_df = main_df.append(current_df)
main_df['date'] = main_df['starttime'].apply (lambda x: x.split(' ')[0])
```

```python
main_df.head()
```

```python
main_df['date'].unique()
```

```python
#Write data to CSV
out_url = os.path.join('..','Raw_Data','JC-ALL-citibike-tripdata.csv')
main_df.to_csv(out_url,index=False)
```
